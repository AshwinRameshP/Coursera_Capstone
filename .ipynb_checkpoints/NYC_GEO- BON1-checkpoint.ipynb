### Part 1 Download and Explore New York city geographical coordinates dataset

Link to the dataset: https://geo.nyu.edu/catalog/nyu_2451_34572 .

There is a total of 5 boroughs and 306 neighborhoods in New York city . To segement the neighborhoods and explore them, we will essentially need the above dataset that contains the 5 boroughs and the neighborhoods that exist in each borough as well as the the latitude and logitude coordinates of each neighborhood. 

Now, let's download all the dependencies that we will need.

```python
# library for data analsysis and using dataframes
import pandas as pd 
#pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows', None)

# convert an address into latitude and longitude values
from geopy.geocoders import Nominatim 

# map rendering library
import folium 

#to get json file in anaconda from url
import requests
print('Libraries imported.')
```

## Load and Explore dataset

```python
newyork_data = requests.get('https://ibm.box.com/shared/static/fbpwbovar7lf8p5sgddm06cgipa2rxpe.json').json()
print('Data downloaded!')
newyork_data
```

All the relevant data is in the *features* key, which is basically a list of the neighborhoods. So, define a new variable that includes this data.

```python
neighborhoods_data = newyork_data['features']
neighborhoods_data
```

#### Tranform the data into a *pandas* dataframe
The next task is essentially transforming this data of nested Python dictionaries into a *pandas* dataframe. Start by creating an empty dataframe.

```python
# define the dataframe columns
column_names = ['Borough', 'Neighborhood', 'Latitude', 'Longitude'] 

# instantiate the dataframe
neighborhoods = pd.DataFrame(columns=column_names)

neighborhoods
```

Then loop through the data and fill the dataframe one row at a time.

```python
for data in neighborhoods_data:
    borough = neighborhood_name = data['properties']['borough'] 
    neighborhood_name = data['properties']['name']
        
    neighborhood_latlon = data['geometry']['coordinates']
    neighborhood_lat = neighborhood_latlon[1]
    neighborhood_lon = neighborhood_latlon[0]
    
    neighborhoods = neighborhoods.append({'Borough': borough,
                                          'Neighborhood': neighborhood_name,
                                          'Latitude': neighborhood_lat,
                                          'Longitude': neighborhood_lon}, ignore_index=True)
neighborhoods.head()
```

Let's make sure that the dataset has all 5 boroughs and 306 neighborhoods.

```python
print('The dataframe has {} boroughs and {} neighborhoods.'.format(
        len(neighborhoods['Borough'].unique()),
        neighborhoods.shape[0]
    )
)
```

```python
neighborhoods.to_csv('NYC_GEO_BON1.csv',index=False)
```

#### Use geopy library to get the latitude and longitude values of New York City.

```python
address = 'New York City, NY'

geolocator = Nominatim(user_agent="Jupyter")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of New York City are {}, {}.'.format(latitude, longitude))
```

#### Create a map of New York with neighborhoods superimposed on top.

```python
# create map of Toronto using latitude and longitude values
map_NewYork = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(neighborhoods['Latitude'], neighborhoods['Longitude'], neighborhoods['Borough'], neighborhoods['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_NewYork)  
    
map_NewYork
```
